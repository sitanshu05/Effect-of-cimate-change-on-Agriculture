```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv(r"C:\Users\sitan\OneDrive\Desktop\4th sem\EDA\cropsOfCountries\Crops_AllData_Normalized.csv",encoding="latin1")
```

```python
countries =['afghanistan','bangladesh',"brazil",'china','colombia','france','germany','india','japan','kenya','mexico','russia','turkey','united states of america','uzbekistan','russian federation']
crops = pd.read_csv(r"C:\Users\sitan\OneDrive\Desktop\4th sem\EDA\project1\project\reducedData\cropRecommendation.csv")
crops = pd.Series(crops['label'])
crops
```

```python
indexDef = df[(df['Year Code'] < 2010) | (~df['Area'].str.lower().isin(countries)) | (~df['Item'].str.lower().isin(crops))].index
indexDef
```

```python
df.drop(indexDef ,axis=0,inplace=True)
df.drop(['Year Code','Flag','Area Code' ,'Item Code', 'Element Code'],axis =1,inplace =True)
df
```

```python
dfHarvest = df[df['Element']=='Area harvested'] #unit ha 
dfHarvest.drop(['Unit' ,'Element'],axis =1,inplace = True)
dfHarvest.to_csv(r"C:\Users\sitan\OneDrive\Desktop\4th sem\EDA\project1\project\reducedData\cropsOfCountries\harvestedCropsDataset.csv")
dfHarvest
```

```python
dfProduction = df[df['Element']=='Production'] # unit tonnes
dfProduction.drop(['Unit' ,'Element'],axis =1,inplace = True)
dfProduction.to_csv(r"C:\Users\sitan\OneDrive\Desktop\4th sem\EDA\project1\project\reducedData\cropsOfCountries\productionCropsDataset.csv")
dfProduction
```

```python
dfYield = df[df['Element']=='Yield'] #unit hg/ha
dfYield.drop(['Unit' ,'Element'],axis =1,inplace = True)
dfYield.to_csv(r"C:\Users\sitan\OneDrive\Desktop\4th sem\EDA\project1\project\reducedData\cropsOfCountries\yieldCropsDataset.csv")
dfYield
```
